# Numpy: Numeric computing library

NumPy (Numerical Python) is one of the core packages for numerical computing in Python. Pandas, Matplotlib, Statmodels and many other Scientific libraries rely on NumPy.

NumPy major contributions are:

* Efficient numeric computation with C primitives
* Efficient collections with vectorized operations
* An integrated and natural Linear Algebra API
* A C API for connecting NumPy with libraries written in C, C++, or FORTRAN.

Let's develop on efficiency. In Python, **everything is an object**, which means that even simple ints are also objects, with all the required machinery to make object work. We call them "Boxed Ints". In contrast, NumPy uses primitive numeric types (floats, ints) which makes storing and computation efficient.

## Boolean arrays
_(Also called masks)_

```python
a = np.arange(4)
```

```python
a
```

```python
a[[0, -1]]
```

```python
a[[True, False, False, True]]
```

```python
a >= 2
```

```python
a[a >= 2]
```

```python
a.mean()
```

```python
a[a > a.mean()]
```

```python
a[~(a > a.mean())]
```

```python
a[(a == 0) | (a == 1)]
```

```python
a[(a <= 2) & (a % 2 == 0)]
```

```python
A = np.random.randint(100, size=(3, 3))
```

```python
A
```

```python
A[np.array([
    [True, False, True],
    [False, True, False],
    [True, False, True]
])]  #Actually doing this takes alot of time, people usually type out the condition and then mask it back to the numpy array.
```

```python
A > 30
```

```python
A[A > 30]
```

## Size of objects in Memory

### Int, floats

```python
# An integer in Python is > 24bytes
sys.getsizeof(1) #Which mean you need 28 bytes to contain an interger 1
```

```python
# Longs are even larger
sys.getsizeof(10**100)
```

```python
np.dtype(np.int8).itemsize
```

```python
# Numpy size is much smaller
np.dtype(int).itemsize
```

```python
np.dtype(float).itemsize
```

### Lists are even larger

```python
# A one-element list
sys.getsizeof([1])
```

```python
# An array of one element in numpy
np.array([1]).nbytes
```

### And performance is also important

```python
l = list(range(1000)) #This is a list (in python)
```

```python
a = np.arange(1000) #This is a numpy array
```

```python
%time np.sum(a ** 2) #It execute with much less time in numpy
```

```python
%time sum([x ** 2 for x in l])
```
