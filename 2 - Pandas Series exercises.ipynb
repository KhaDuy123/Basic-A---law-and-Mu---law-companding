## Series boolean arrays (also called masks)

### Given the X pandas Series, make a mask showing negative elements


```python
# your code goes here
X = pd.Series([32,22,100,354], index = ['First','Second','Third','Fourth'], name = 'Caffe value')
print(X)
mask = X <0
print(X[mask])
print(mask)
```

```python
X = pd.Series([-1,2,0,-4,5,6,0,0,-9,10])

mask = X <= 0
mask
```

![green-divider](https://user-images.githubusercontent.com/7065401/52071924-c003ad80-2562-11e9-8297-1c6595f8a7ff.png)

### Given the X pandas Series, get the negative elements


```python
# your code goes here
X = pd.Series([-2,3,-5,20],index = ['First','Second','Third','Fourth'], name = 'TEXT')
print(X)
mask = X < 0
print(X[mask])
```

```python
X = pd.Series([-1,2,0,-4,5,6,0,0,-9,10])

mask = X <= 0
X[mask]
```

### Given the X pandas Series, get numbers higher than 5


```python
# your code goes here
X = pd.Series([10,20,6,30], index = ['First','Second','Third','Fourth'], name = 'Number')
print(X)
mask = X > 5
print(X[mask])
```

### Given the X pandas Series, get numbers higher than the elements mean

```python
# your code goes here
X = pd.Series([1,2,3,4,5,6], index = ['One','Two','Three','Four','Five','Six'], name = 'Number ABC')
print(X, "\n")
mask = X > X.mean()
print(X.mean())
print(X[mask])
```

```python
X = pd.Series([-1,2,0,-4,5,6,0,0,-9,10])

mask = X > X.mean()
X[mask]
```

### Given the X pandas Series, get numbers equal to 2 or 10


```python
# your code goes here
X = pd.Series([2,10,6,9,10,3,2], index = ['One','Two','Three','Four', 'Five','Six','Seven'], name = 'Number ABC')
mask = (X == 2) | (X == 10)
print(X[mask])
```

```python
X = pd.Series([-1,2,0,-4,5,6,0,0,-9,10])

mask = (X == 2) | (X == 10)
X[mask]
```
